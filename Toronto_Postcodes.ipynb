```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup # Scraping using Beautiful Soup

req = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M")

soup = BeautifulSoup(req.content,'lxml')

table = soup.find_all('table')[0]

df = pd.read_html(str(table)) # Making readable in pandas

neighborhood=pd.DataFrame(df[0])

neighborhood
```

```python
neighborhood.replace("Not assigned", np.nan, inplace = True) # relabling all missing data 
neighborhood.head(5)
```

```python
neighborhood.dropna(subset=["Borough"], axis=0, inplace=True) # Dropping cells where borough isn't defined
neighborhood.reset_index(drop=True, inplace=True)
neighborhood
```

```python
neighborhood['Neighbourhood'].replace(np.nan, 'Borough', inplace=True) # Replacing an undefined neighborhood with its borough
neighborhood
```

```python
Test = neighborhood.groupby(by='Postcode', axis=0, level=None, as_index=True, sort=True, group_keys=True, squeeze=False)
# Sorting by Postcode
```
